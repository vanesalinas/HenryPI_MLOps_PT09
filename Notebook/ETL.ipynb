Creamos una funcion para desanidar las columnas excepto 'belongs to collection' ya que le daremos otro tratamiento a la misma. En el resto de la columnas, su estructura es un identificador y un nombre dentro de cada diccionario que contiene la lista; almacenaremos solo el campo 'name' dentro de una nueva lista. Esto es para no extender demasiado el dataframe y ademas porque considero que los ID de estos campos no son demasiado relevantes para nuestro analisis; y si tuviesemos que realizar una busqueda entre los generos de una pelicula por ejemplo, lo podemos hacer mediante el nombre.

```python
def get_names(columna):
  """
  Extrae los nombres dentro del campo 'name' de una lista de diccionarios
  y asigna 'Desconocido' si el campo no existe

  Parametros:
    columna: Nombre de la columna a desanidar.

  Retorno:
    Lista de nombres.
  """

  # Manejar valores NaN
  if pd.isna(columna):
    return 'Desconocido'

  try:
    # Convertir la columna a listas de diccionarios si es necesario
    columna = ast.literal_eval(columna)

    if not columna:
      return 'Desconocido'  # Retornar 'Desconocido' para listas vacias
    return [item['name'] for item in columna]  #extraer los nombres de los géneros
  except (ValueError, SyntaxError,TypeError):
        # Manejar errores de evaluación
        return 'Error en evaluación'
```

Creamos una copia de df_movies para no modificar el dataframe original y evaluamos cada columna con la funcion creada

```python
df = df_movies.copy()
```

Aplicamos la funcion en las columnas 'genres', 'production_companies', 'production_countries' y 'spoken_languages' creando una nueva columna para almacenar los valores.
Luego eliminamos la columna original

```python
# Aplicar la funcion en la columna 'genres' y crear una nueva columna
df['genres_names'] = df['genres'].apply(get_names)

# Eliminar la columna original 'genres'
df.drop('genres', axis=1, inplace=True)
```

```python
df['production_companies_names'] = df['production_companies'].apply(get_names)
df.drop('production_companies', axis=1, inplace=True)
```

```python
df['spoken_languages_names'] = df['spoken_languages'].apply(get_names)
df.drop('spoken_languages', axis=1, inplace=True)
```

```python
df['production_countries_names'] = df['production_countries'].apply(get_names)
df.drop('production_countries', axis=1, inplace=True)
```

Ahora separaremos los campos de la columna 'belongs_to_collection'.

```python
def get_collection(coleccion):
    if pd.isna(coleccion):
        return pd.Series()
    elif isinstance(coleccion, str): # Verificar si coleccion es una cadena
        try:
            # Intentar convertir la cadena a un diccionario
            coleccion = eval(coleccion)
        except:
            return pd.Series() # Manejar casos donde la conversión falla

    if isinstance(coleccion, dict):
        return pd.Series({f'belongs_to_collection_{k}': v for k, v in coleccion.items()})
    else:
        return pd.Series()

# Aplicar la función y crear un nuevo DataFrame
collection_df = df['belongs_to_collection'].apply(get_collection)

# Concatenar el nuevo DataFrame con el original
df = pd.concat([df, collection_df], axis=1)
```

```python
df.drop('belongs_to_collection', axis=1, inplace=True)
```

En ambas hay mucha informacion irrelevante, lo cual al expandir los datos nos generaria dataframes muy extensos; por lo que solo vamos a extraer el nombre de los actores, los directores y el respectivo id de pelicula de cada uno ya que esta es la informacion necesaria para nuestro analisis.

```python
df_credits['crew'] = df_credits['crew'].apply(ast.literal_eval)
```

```python
df_credits['cast'] = df_credits['cast'].apply(ast.literal_eval)
```

```python
# Crear listas vacías para almacenar datos expandidos
directors = []
actors = []

# Iterar sobre cada película en el DataFrame
for index, row in df_credits.iterrows():
    movie_id = row['id']
    cast_list = row['cast']
    crew_list = row['crew']

    # Iterar sobre el elenco y el equipo de la película
    for member in cast_list:
        actor_data = {
          'id': movie_id,
          'actor_name': member['name'],
        }
        actors.append(actor_data)

    for member in crew_list:
        if member['job'] == 'Director':
            director_data = {
                'id': movie_id,
                'director_name': member['name'],
            }
            directors.append(director_data)

# Convertir las listas a DataFrames
df_directors = pd.DataFrame(directors)
df_actors = pd.DataFrame(actors)
```

## Return

Se crea la columna con el retorno de inversión, llamada return dividiendo 'revenue' por 'budget'; cuando no hay datos disponibles para calcularlo, tomara el valor 0.

Primero verificamos el tipo de datos de cada columna para saber si son numericos y asi poder efectuar operaciones aritmeticas.

```python
df['revenue'].dtype
```

```python
df['budget'].dtype
```

```python
# Convertir la columna 'budget' a números
df['budget'] = pd.to_numeric(df['budget'], errors='coerce')

# Calcular el retorno de inversión
df['return'] = df.apply(lambda row: round(row['revenue'] / row['budget'],2) if row['budget'] != 0 else 0, axis=1)

df.head()
```
